### CSV data modification

```python
ori_csv_path = 'time_series_covid19_confirmed_global.csv'
mod_csv_path = 'time_series_covid19_confirmed_global_simplified.csv'
```

```python
with open(ori_csv_path, 'r') as f:
    csv_reader = csv.reader(f)
    rows = list(csv_reader)

len(rows)
```

```python
dates = rows[0][4:]
for i in range(len(dates)):
    if dates[i].startswith('2002'):
        # Some dates are wrong, e.g. 2002/1/20 should be 2/1/20
        dates[i] = dates[i].replace('2002', '2')

dates
```

```python
for row in rows:
    if row[0] == '' and row[1] == 'US':
        US_data = row[4:]
        break

US_data
```

```python
with open(mod_csv_path, 'w') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow(['Dates', 'Confirmed Cases'])
    for d, c in zip(dates, US_data):
        csv_writer.writerow([d, c])
```
